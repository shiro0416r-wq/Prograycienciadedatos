```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Dannap7337/Pato/main/Base_limpiaclase.csv')
df=df.drop(columns=['Unnamed: 0'])
df
```

```python
#KDE para balance
plt.Figure(figsize=(10,4))
sns.kdeplot(df['Balance'], fill=True, color="purple")
sns.set_style('dark')
plt.title("Grafica de densidad balance")
plt.show()
```

```python
plt.Figure(figsize=(10,4))
sns.kdeplot(df['Puntos'], fill=True, color="pink")
plt.title("Grafica de densidad de puntos")
plt.show()
```

```python
plt.Figure(figsize=(10,4))
sns.kdeplot(df['Salario'], fill=True, color="green")
plt.title("Grafica de densidad Salario")
plt.show()
```

```python
plt.Figure(figsize=(10,4))
sns.kdeplot(df['CreditScore'], fill=True, color="yellow")
plt.title("Grafica de densidad Score crediticio")
plt.show()
```

```python
fig, axes = plt.Subplot(2, 2, 2)
sns.kdeplot(df["Balance"], fill=True, ax=axes[0,0], color="pink")
axes[0,0].set_title("Distribucion de balace")

sns.kdeplot(df["Salario"], fill=True, ax=axes[0,1], color="purple")
axes[0,1].set_title("Distribucion de salario")

sns.kdeplot(df["CreditScore"], fill=True, ax=axes[1,0], color="yellow")
axes[1,0].set_title("Distribucion de CreditScore")

sns.kdeplot(df["Edad"], fill=True, ax=axes[1,1], color="red")
axes[1,1].set_title("Distribucion de Edad")

plt.tight_layout()

plt.show()
```

```python
gender_counts = df["Genero"].value_counts()

plt.figure(figsize=(12,12))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90,
        colors=['purple', 'pink'])
plt.title("Distribucion por genero")
plt.show()
```

```python
sns.violinplot(x='CreditScore', data=df)

plt.title("Distribucion de CreditScore")
plt.show()
```
