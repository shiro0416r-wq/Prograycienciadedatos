### Limpieza base de datos (Parte 2) ###

```python
import numpy as np
```

```python
import pandas as pd
df = pd.read_csv("df_semilimpiio.csv")
df
```

```python
#borramos una columna que aÃ±adi por accidente
df = df.drop(columns="JP_Sales\t")
```

### Limpieza de columna "Platform" ###

En este caso contamos con 556 datos nulos en la columna de plataforma. Sin embargo al ser una columna NECESARIA para este analisis Y ademas representar menos del 5% de los datos (2.99%). Se opto por simplemente dropear 
esos datos nulos

```python
df = df.dropna(subset=["Platform"])
df
```

### Limpieza columna "Year" ###

De la misma manera que la columna anterior, al ser esta importante para mantener la integridad del analisis Y representar menos del 5% de mis datos (4,7%), se opto por dropearlos

```python
df["Year"]= df["Year"].replace("Auto%#", np.nan)
df = df.dropna(subset=["Year"])
df
```

### Eliminando registros de las columnas "ventas" con mas de 1 NaN ###

En este momento lo correcto seria rellenar los registros que solamente tienen "NaN" en las columna "Global_sales", esto pudiendo ser facilmente calculado mediante la suma de las ventas por region. Sin embargo antes de hacer esto
primero debemos deshacernos de los registros que cuenten con mas de 1 "NaN" en las demas columnas de ventas. Por que? porque de esta forma podemos calcular los registros que solo tengan un NaN mediante una ecuacion simple (por ejemplo: global_sales = NA sales + JP sales + EU sales + Other sales) mientras que en caso contrario (tener por ejemplo NaN en "jp sales y en eu sales") no se podria calcular correctamente.

```python
#1.- Creamos una lista con las columnas de ventas
columnas_ventas = ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']
```

```python
#2.- Creamos una columna que contara cuantos datos NaN tiene un registro en las columnas de ventas
df["NaN_Counts"] = df[columnas_ventas].isna().sum(axis=1)
```

```python
#3.- Ahora conservamos solo los registros que contengan 1 o menos datos NaN en las columnas de ventas
df = df[df['NaN_Counts'] <= 1]
```

```python
#4.- Nos deshacemos de la columna que nos auxilio
df = df.drop(columns="NaN_Counts")
```

```python
#Dropeamos los duplicados (cosa que no se habia hecho hasta ahora)
df.drop_duplicates()
```

```python
#comprobamos cuantos datos nulos tenemos
df.isna().sum()
```

### Calculo de datos nulos en columnas de ventas ###

Una vez que ya solamente nos quedamos con los registros de ventas de los cuales podemos calcular cualquiera de sus columnas en caso de tener datos NaN, procedemos a calcularlos.

```python
#1.- Crear una lista con los registros NaN de global sales
global_nan = df["Global_Sales"].isna()
```

```python
#2.- Tomamos los registros nulos con .loc y aplicamos la formula
#(Se aÃ±ade .loc a cada uno de los registros a sumarse para aegurarnos que solo se sumen los registros de una misma fila)
df.loc[global_nan, "Global_Sales"] = (
    df.loc[global_nan, 'EU_Sales'] +
    df.loc[global_nan, 'JP_Sales'] +
    df.loc[global_nan, 'NA_Sales'] +
    df.loc[global_nan, 'Other_Sales']
)
```

```python
#3.- Procedemos a hacer lo mismo  con el resto de las columnas de ventas
eu_nan = df["EU_Sales"].isna()

df.loc[eu_nan, "EU_Sales"] = (
    df.loc[eu_nan, "Global_Sales"]-
    df.loc[eu_nan,"Other_Sales"]-
    df.loc[eu_nan,"JP_Sales"]-
    df.loc[eu_nan, "NA_Sales"]
)
```

```python
jp_nan = df["JP_Sales"].isna()

df.loc[jp_nan, "JP_Sales"] = (
    df.loc[jp_nan, "Global_Sales"]-
    df.loc[jp_nan,"Other_Sales"]-
    df.loc[jp_nan,"EU_Sales"]-
    df.loc[jp_nan, "NA_Sales"]
)
```

```python
na_nan = df["NA_Sales"].isna()

df.loc[na_nan, "NA_Sales"] = (
    df.loc[na_nan, "Global_Sales"]-
    df.loc[na_nan,"Other_Sales"]-
    df.loc[na_nan,"EU_Sales"]-
    df.loc[na_nan, "JP_Sales"]
)
```

```python
other_nan = df["Other_Sales"].isna()

df.loc[other_nan, "Other_Sales"] = (
    df.loc[other_nan, "Global_Sales"]-
    df.loc[other_nan,"NA_Sales"]-
    df.loc[other_nan,"EU_Sales"]-
    df.loc[other_nan, "JP_Sales"]
)
```

```python
#Ahora corroboramos cuantos datos nulos tenemos
df.isnull().sum()
```

Para finalizar con esta limpieza, al igual que con la columna anterior, simplemente dropearmos los NaN por las mismas razones (Ser categorica y representar menos del 5% de los datos (3.1%))
