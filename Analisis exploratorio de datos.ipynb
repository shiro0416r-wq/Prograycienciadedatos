```python

from sklearn.model_selection import train_test_split

#X sera las variables que usaremos para tratar de predecir nuestra variable onbjetivo
x = df[["Ventas_NA", "Ventas_EU", "Ventas_JP", "Ventas_Otras", "Género", "Año"]]

#Nuestro modelo trabaja con datos numericos, por lo tanto convertimos
#la columna genero a formato numerico "one-hot"
x = pd.get_dummies(x, columns=['Género'], drop_first=True)

#Creamos nuestra variable objetivo
y = df["Ventas_Globales"]

```

```python
#Importamos el modelo
from sklearn.linear_model import LinearRegression


#El modelo aprendera en base a nuestros datos
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
reg_lineal = LinearRegression()
reg_lineal.fit(x_train, y_train)
```

```python
#Importamos el segundo modelo que usaremos
from sklearn.ensemble import RandomForestRegressor

#El modelo aprendera en base a nuestros datos
reg_rf = RandomForestRegressor(random_state=42)
reg_rf.fit(x_train, y_train)
```

```python
#Hacemos predicciones con nuestro primer modelo
y_pred_lineal = reg_lineal.predict(x_test)
y_pred_rf = reg_rf.predict(x_test)
```

```python
from sklearn.model_selection import GridSearchCV

#Buscamos la mejor combinacion de parametros para nuestro segundo modelo
parametros = {"n_estimators": [50, 100, 200], "max_depth": [None, 18, 20]}
grid = GridSearchCV(RandomForestRegressor(), parametros, cv=3)
grid.fit(x_train, y_train)
best_rf = grid.best_estimator_
```

```python
#Evaluamos nuestro modelo de regresion lineal segun las metricas adecuadas

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
y_pred_lineal = reg_lineal.predict(x_test)

mae_lineal = mean_absolute_error(y_test, y_pred_lineal)
mse_lineal = mean_squared_error(y_test, y_pred_lineal)
rmse_lineal = np.sqrt(mse_lineal)
r2_lineal = r2_score(y_test, y_pred_lineal)

#Se muestran los resultados obtenidos
print("Resultados/Evaluacion de la Regresión Lineal:")
print(f"MAE: {mae_lineal}")
print(f"RMSE: {rmse_lineal}")
print(f"R2: {r2_lineal}")
```

```python
#Evaluamos nuestro modelo de RF segun las metricas adecuadas

y_pred_rf = best_rf.predict(x_test)

mae_rf = mean_absolute_error(y_test, y_pred_rf)
mse_rf = mean_squared_error(y_test, y_pred_rf)
rmse_rf = np.sqrt(mse_rf)
r2_rf = r2_score(y_test, y_pred_rf)

print("Resultados/Evaluacion de Random Forest:")
print(f"MAE: {mae_rf}")
print(f"RMSE: {rmse_rf}")
print(f"R2: {r2_rf}")
```

```python
#Grafico de dispersion entre valores reales y predichos de regresion lineal
plt.scatter(y_test, y_pred_lineal) 
plt.xlabel('Valores reales') 
plt.ylabel('Predicciones') 
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2, label='Predicción Perfecta')
plt.title('Predicciones vs Valores reales')

plt.show()
```

```python
#Grafico de dispersion entre valores reales y predichos de RF
plt.scatter(y_test, y_pred_rf) 
plt.xlabel('Valores reales') 
plt.ylabel('Predicciones') 
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2, label='Predicción Perfecta')
plt.title('Predicciones vs Valores reales')

plt.show()
```
