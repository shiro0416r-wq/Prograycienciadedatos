### Limpieza base de datos ###

```python
#libreria numpy
import numpy as np
```

```python
#Libreria pandas
import pandas as pd
df = pd.read_csv("df_sucio.csv")
df
```

### Limpieza columna "Rank" ###

Para la limpieza de esta columna, al ser un dato no vital para el analisis, procedemos simplemente a rellenar los datos NaN con el numero 0 en este caso

```python
#Rellenamos los NaN de la columna Rank con ceros
df["Rank"] = df["Rank"].fillna(0)
df
```

```python
#Comprobamos que efectivamente, ya no tenemos datos nulos en esa columna
df.isnull().sum()
```

Una vez que nos deshicimos de los datos NaN, tenemos que comprobar que no hayan datos atipicos o erroneos

```python
#Aqui logramos ver que hay 381 datos invalidos (Auto%#)
df["Rank"].value_counts()
```

```python
#Para limpiarlos, simplemente reemplazamos los datos invalidos por datos NaN mediante Numpy
df["Rank"]= df["Rank"].replace("Auto%#", np.nan)
df
```

```python
#Una vez que los datos ya son NaN, procedemos a reemplazarlos con el numero 0
df["Rank"] = df["Rank"].fillna(0)
df
```

### Limpieza columna "Name" ###

Para la limpieza de esta columna se decidio dropear tanto los datos nulos como los datos invalidos. Esto debido a varios factores:

1.- Ser una columna que no puede llenarse con un dato neutro al ser de suma importancia para el analisis

2.- Ser una columna del tipo categorica

3.- Los datos de esta columna no se pueden inferir mediante los demas datos en la base

4.- Rellenar con datos inciertos podria poner en peligro la integridad del analisis

```python
#Comprobamos que en efecto, hayan datos nulos
df["Name"].unique()
```

```python
#Corroboramos cuantos datos nulos hay
df["Name"].isnull().sum()
```

```python
#Procedemos a dropear esos registros
df = df.dropna(subset=["Name"])
```

```python
#Reemplazamos los datos invalidos por NaN
df["Name"]= df["Name"].replace("Auto%#", np.nan)
```

```python
#Volvemos a eliminar los nuevos datos NaN
df = df.dropna(subset=["Name"])
```

```python
#Comprobamos que ya no haya datos nulos
df["Name"].isnull().sum()
```
